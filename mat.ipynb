```python
import scipy.io
import matplotlib.pyplot as plt

# .mat 파일 로드
mat_file_path = '/media/NAS/DATASET/DFWild-Cup/noiseprint/train/real/train_real_0000048.mat'
data = scipy.io.loadmat(mat_file_path)

# 파일 내용 확인 (데이터 구조를 확인하여 필요한 변수 선택)
print(data.keys())  # 변수 목록 출력

# dict_keys(['__header__', '__version__', '__globals__', 'noiseprint', 'QF', 'time'])
# 특정 변수를 시각화 (예: 데이터가 2D 배열인 경우)
# data['your_variable_name']으로 접근 (예시용 변수명 사용)
variable_data = data['noiseprint']

# 2D 데이터 예시 시각화
plt.imshow(variable_data, cmap='viridis', vmin=-1, vmax=1)  # 'seismic' 컬러맵으로 -1에서 1 범위 강조
plt.colorbar()
plt.title('Noiseprint (Emphasized 0 to 1)')
plt.show()
```
